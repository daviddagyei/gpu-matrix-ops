# Modular Trade Notebook

This notebook runs the registry-driven trading engine and keeps every output in pandas DataFrame form.


```python
import pandas as pd

from engine import run_engine_bundle

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
```

```python
submit_orders = False

engine_config = {
    "preset": "alpaca_sma_default",
    "symbol": "BTCUSD",
    "timeframe": "1Hour",
    "lookback_days": 7,
    "qty": 0.001,
    "submit_orders": submit_orders,
    "market_data_provider": "alpaca",
    "strategy": "sma_trend",
    "strategy_params": {
        "sma_window": 20,
    },
    "risk_rules": ["max_position", "no_duplicate_entry"],
    "risk_params": {
        "max_position_qty": 1,
    },
    "execution_handler": "alpaca_paper",
}

pd.DataFrame([engine_config])
```

```python
bundle = run_engine_bundle(engine_config)

settings_df = bundle["settings_df"]
market_data_df = bundle["market_data_df"]
signal_df = bundle["signal_df"]
risk_df = bundle["risk_df"]
execution_df = bundle["execution_df"]
positions_df = bundle["positions_df"]
orders_df = bundle["orders_df"]
summary_df = bundle["summary_df"]
```
